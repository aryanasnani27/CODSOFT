```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x_test=advertisement[['TV','Radio','Newspaper']]
y_test=advertisement['Sales']
```

```python
import statsmodels.api as sm
```

```python
x1_test=sm.add_constant(x_test)
reg=sm.OLS(y_test,x1_test)
model=reg.fit()
print(model.summary())
```

```python
residual=model.resid
residual
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_test,y_test,test_size=0.2,random_state=22)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
y_predict
```
